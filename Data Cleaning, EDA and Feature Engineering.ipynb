```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#read golf csv (with stats from 2010 to 2018)
df = pd.read_csv('PGA_Data_Historical.csv')
```

```python
df.head()
```

```python
#change csv format from long to wide
df = df.set_index(['Player Name', 'Variable', 'Season'])['Value'].unstack('Variable').reset_index()


```

```python
df.head()
```

```python
#remove apostrophes from columns
df.columns = df.columns.str.replace("'"," ")
```

```python
df.head()
```

```python
#drop rows without scoring average
df = df.dropna(how='any', subset=['Stroke Differential Field Average - (PLAYER S SCORE)'])
```
